```python
with open("hostel_bois.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

print("Total lines:", len(lines))
```

```python
for i in range(5):
  print(lines[i])
```

```python
messages=[]
```

```python
for line in lines:
    line = line.strip()

    if line == "":
        continue

    if " - " not in line:
        continue

    parts = line.split(" - ", 1)

    timestamp = parts[0]
    data = parts[1]

    if ": " not in data:
        continue

    sender, text = data.split(": ", 1)

    message = {
        "timestamp": timestamp,
        "sender": sender,
        "text": text
    }

    messages.append(message)
```

```python
print("Total parsed messages:", len(messages))
print(messages[:5])
```

```python
person_count = {}

for msg in messages:
    name = msg["sender"]

    if name in person_count:
        person_count[name] += 1
    else:
        person_count[name] = 1

print(person_count)
```

```python
sorted_people = sorted(
    person_count.items(),
    key=lambda x: x[1],
    reverse=True
)

for name, count in sorted_people:
    print(name, count)
```

```python
word_count = {}

stop_words = [
    "i", "is", "the", "a", "and", "or", "to",
    "of", "in", "on", "for", "it", "this"
]

for msg in messages:
    text = msg["text"].lower()

    words = text.split()

    for word in words:
        word = word.strip(".,!?()[]{}\"'")

        if word not in stop_words and word != "":
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1
```

```python
top_words = sorted(
    word_count.items(),
    key=lambda x: x[1],
    reverse=True
)

print("TOP 10 WORDS")
print("-" * 30)

for word, count in top_words[:10]:
    print(f"{word:<10} : {count}")
```

```python
import numpy as np
```

```python
participants = []

for msg in messages:
    name = msg["sender"]

    if name not in participants:
        participants.append(name)

print(participants)
```

```python
heatmap = np.zeros((len(participants), 24), dtype=int)

print(heatmap)
```

```python
for msg in messages:
    name = msg["sender"]

    time = msg["timestamp"].split(", ")[1]
    hour = int(time.split(":")[0])

    row = participants.index(name)

    heatmap[row][hour] += 1
```

```python
print(heatmap)
```

```python
print("ACTIVITY HEATMAP")
print("      ", end="")

for hour in range(24):
    print(f"{hour:02}", end=" ")

print()

for i, person in enumerate(participants):
    print(f"{person:<6}", end=" ")

    for value in heatmap[i]:
        print(f"{value:02}", end=" ")

    print()
```

```python
from datetime import datetime, timedelta
```

```python
for msg in messages:
    msg["datetime"] = datetime.strptime(
        msg["timestamp"],
        "%d/%m/%y, %H:%M"
    )
```

```python
response_times = {}

for i in range(1, len(messages)):
    current = messages[i]
    previous = messages[i - 1]

    if current["sender"] != previous["sender"]:

        gap = (
            current["datetime"] -
            previous["datetime"]
        ).total_seconds() / 60

        name = current["sender"]

        if name not in response_times:
            response_times[name] = []

        response_times[name].append(gap)
```

```python
average_response = {}

for person, times in response_times.items():
    average_response[person] = sum(times) / len(times)

for person, avg in average_response.items():
    print(person, ":", round(avg, 2), "minutes")
```

```python
fastest = min(
    average_response,
    key=average_response.get
)

slowest = max(
    average_response,
    key=average_response.get
)

print("Fastest replier:", fastest)
print(
    "Average time:",
    round(average_response[fastest], 2),
    "minutes"
)

print("Slowest replier:", slowest)
print(
    "Average time:",
    round(average_response[slowest], 2),
    "minutes"
)
```

```python
person_days = {}

for msg in messages:
    name = msg["sender"]
    day = msg["datetime"].date()

    if name not in person_days:
        person_days[name] = set()

    person_days[name].add(day)

print(person_days)
```

```python
all_days = set()

for msg in messages:
    all_days.add(msg["datetime"].date())

total_days = len(all_days)

for person in person_days:
    silent_days = total_days - len(person_days[person])

    print(person, "was silent for", silent_days, "days")
```

```python
burst_count = {}
current_sender = ""
count = 0

for msg in messages:
    sender = msg["sender"]

    if sender == current_sender:
        count += 1
    else:
        if current_sender != "":
            burst_count[current_sender] = max(
                burst_count.get(current_sender, 0),
                count
            )

        current_sender = sender
        count = 1

print("Spammer scores:", burst_count)
```

```python
night_count = {}
total_count = {}

for msg in messages:
    name = msg["sender"]
    hour = msg["datetime"].hour

    total_count[name] = total_count.get(name, 0) + 1

    if hour >= 23 or hour <= 4:
        night_count[name] = night_count.get(name, 0) + 1

for person in total_count:
    percentage = (night_count.get(person, 0) /
                  total_count[person]) * 100

    print(person, "Night %:", round(percentage, 2))
```

```python
caring_words = [
    "okay", "safe", "eat",
    "sleep", "please",
    "take care"
]

care_score = {}

for msg in messages:
    text = msg["text"].lower()

    for word in caring_words:
        if word in text:
            name = msg["sender"]

            care_score[name] = (
                care_score.get(name, 0) + 1
            )

print(care_score)
```

```python
length_data = {}

for msg in messages:
    name = msg["sender"]

    words = len(msg["text"].split())

    if name not in length_data:
        length_data[name] = []

    length_data[name].append(words)

for person, data in length_data.items():
    average = sum(data) / len(data)

    print(person,
          "Average words:",
          round(average, 2))
```

```python
drama_score = {}

for msg in messages:
    text = msg["text"]

    if text.isupper() or text.count("!") >= 2:
        name = msg["sender"]

        drama_score[name] = (
            drama_score.get(name, 0) + 1
        )

print(drama_score)
```

```python
ghost_score = {}

for person in person_days:
    ghost_score[person] = (
        total_days - len(person_days[person])
    )

print(ghost_score)
```
